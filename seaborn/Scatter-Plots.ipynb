```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/dm_office_sales.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
sns.scatterplot(x='salary', y='sales', data=df);
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='salary', y='sales', hue='division')
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='salary', y='sales', hue='work experience')
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='salary', y='sales', hue='work experience', palette='viridis')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)

# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
```
