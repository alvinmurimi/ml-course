# Groupby Operations and Multi-level Index

A `groupby()` operation allows us to examine data on a **per category** basis

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/mpg.csv')
```

```python
df
```

```python
df.dtypes
```

```python
horse_name = df.groupby(['horsepower', 'name']).mean()
```

```python
horse_name
```

```python
horse_name.index
```
